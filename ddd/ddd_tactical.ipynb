# Domain, Subdomain and Bounded Context Helper

This section of the Jupyter Notebook is dedicated to assisting architects and developers in the design and documentation of microservices architectures. By leveraging the capabilities of Generative Pre-trained Transformer (GPT), we aim to streamline the process of identifying potential microservices based on given domain information. This tool is especially useful in the early stages of software development and architecture planning, providing insights and suggestions that can help shape your microservices strategy.

## How It Works

1. **Domain Input**: Users start by specifying the high-level domain of their software project. The domain represents the overarching area within which the software solution operates, such as finance, healthcare, e-commerce, etc.

2. **Subdomain Identification**: Within the specified domain, users then identify subdomains. Subdomains are more specific areas of focus within the general domain, representing distinct segments of the business model or areas of functionality. For example, within the e-commerce domain, subdomains might include inventory management, payment processing, and user experience.

3. **Bounded Contexts Specification**: For each subdomain, users define bounded contexts. A bounded context is a design concept from Domain-Driven Design (DDD) that provides a framework for making decisions about where models apply and how they interact. It helps in clearly defining the limitations and responsibilities of each microservice.

4. **Microservices Suggestion**: Based on the input domain, subdomains, and bounded contexts, GPT will analyze the information and suggest a list of potential microservices. This list includes suggested names and responsibilities of each microservice, aligning with the defined contexts and aiming to cover the identified functionalities.




```python
from dotenv import load_dotenv
load_dotenv()

from openai import OpenAI
client = OpenAI()

def get_completion(prompt, model="gpt-4-1106-preview"):
   messages = [{"role": "user", "content": prompt}]
   response = client.chat.completions.create(model=model,
   messages=messages,
   temperature=0)
   return response.choices[0].message.content

ddd_tactical = {
    "guides_4_microservices" : {
        "srp" : "Single Responsibility Principle (SRP): Apply SRP to microservices. A microservice should do one thing well and have a single reason to change.",
        "manageable" : "Keep microservices small and manageable. However, small is relative and context-dependent; focus on cohesion and autonomy rather than size." ,
        "ddd" : "DDD principles to identify bounded contexts. Each bounded context corresponds to a specific domain model and has explicit boundaries, making it a candidate for a separate microservice."
    }
}

domain = """
  The Payment Gateway domain encompasses the infrastructure and services that facilitate online payment processing between merchants and customers. It acts as an intermediary, securely transmitting payment information from the customer to the merchant's bank account or payment processor. This domain includes handling authorization of transactions, encryption of payment details, fraud detection, and compliance with financial standards and regulations. Payment gateways are essential for e-commerce, enabling businesses to accept various forms of digital payments (such as credit cards, bank transfers, and digital wallets) while ensuring transaction security and integrity.
"""

subdomain = """Fraud Detection and Risk Management this subdomain is responsible for identifying and mitigating fraudulent activities. It employs various tools and techniques, such as machine learning algorithms, behavior analysis, and risk scoring systems, to detect suspicious transactions and prevent fraud."""

bounded_context = "This bounded context is dedicated to the detection and prevention of fraudulent activities. It uses machine learning, behavior analysis, and risk scoring to identify and block suspicious transactions, thereby reducing the risk of fraud."


microservices_prompt = f"""

    Your task is to perform the following actions:
        1. Analyze the domain, subdomain, and bounded context as input
        2. Define the microservices for {bounded_context} using domain and subdomain as a context  
        3. Identify the microservice following the guides.
        4. Give me the microservices list and description of each one. 
        
    
    Domain: {domain}
    Subdomain: {subdomain}
    Bounded context: {bounded_context}
    
    Guides: ddd_tactical{["guides_4_microservices"]}
    
"""

microservices_list = get_completion(microservices_prompt)
print(microservices_list)

```

# Tactical DDD Design Assistant

Following the identification of potential microservices in the previous section, this part of the Jupyter Notebook aids in the deeper architectural design within a selected bounded context. Utilizing the advanced capabilities of Generative Pre-trained Transformer (GPT), we delve into the tactical aspects of Domain-Driven Design (DDD), including the definition of Aggregates, Entities, and Value Objects. This approach ensures a robust and scalable microservice architecture by aligning closely with DDD principles.

## Overview

Domain-Driven Design (DDD) is a software design philosophy that focuses on modeling software to match a complex domain. While the strategic design helps in defining the high-level structure and boundaries, the tactical design is concerned with the finer details within these boundaries. This section assists in identifying:

- **Aggregates**: A cluster of domain objects that can be treated as a single unit. An aggregate has a root and a boundary defined around it.
- **Entities**: Objects that are not defined by their attributes, but rather by a thread of continuity and identity.
- **Value Objects**: Objects that describe certain aspects of the domain with no conceptual identity, fully described by their attributes.

## Process

1. **Bounded Context Selection**: Begin by selecting a bounded context for which you wish to define the tactical design elements. This should be one of the contexts identified in the previous section.

2. **Input Specifications**: Provide detailed descriptions and requirements specific to the selected bounded context. This might include key functionalities, essential operations, and any known constraints or rules governing the domain logic.

3. **Generate DDD Elements**: Based on the provided specifications, GPT will analyze the bounded context and generate a suggested list of Aggregates, Entities, and Value Objects. This list includes names and descriptions, highlighting how they could be implemented to encapsulate the domain logic effectively.




```python
case_management_service = """
    Case Management Service
    Description: Once a suspicious transaction is identified, this service manages the investigation process. It tracks the status of the case, assigns it to investigators, and stores all relevant information until the case is resolved.
"""

service_detail_prompt = f"""
    Your task is to analyse the bounded context in detail and generate the following :
        1. Using Domain Driven Design techniques, Identify the Entities, Value Objects with list of attributes and behaviours for each.
        2. List the services, repositories, etc. required for the application. 

    Bounded Context: {case_management_service}
"""

service_data = get_completion(service_detail_prompt)
print(service_data)

```

# Class Diagram Generator for DDD Artifacts

In this section, we leverage the power of Generative Pre-trained Transformer (GPT) to transform the tactical DDD design artifacts identified in the previous step into comprehensive class diagrams. This automated process aids architects and developers in visualizing the relationships and structures of Aggregates, Entities, and Value Objects within the selected bounded context, facilitating a clearer understanding and communication of the domain model.

## Purpose

Class diagrams are a fundamental part of object-oriented design, offering a visual representation of a system's classes, their attributes, methods, and the relationships among classes. For a microservice architecture, particularly one designed following DDD principles, class diagrams serve as a crucial blueprint that guides the development process, ensuring that the implementation aligns with the domain model.

## How It Works

1. **Review DDD Artifacts**: Begin by reviewing the Aggregates, Entities, and Value Objects identified for your selected bounded context. Ensure that these elements are well-defined and accurately represent the domain model.

2. **Class Diagram Generation**: Input the detailed descriptions of your DDD artifacts into the provided code cell. Upon execution, GPT will analyze these descriptions and generate a class diagram. This diagram will include:
   - Classes for each Entity and Value Object, with attributes and methods.
   - Aggregates represented with their root Entities and contained Value Objects and Entities, highlighting boundaries.
   - Relationships between classes, such as associations, aggregations, and compositions.

3. **Visualization**: The generated class diagram will be visualized directly within the Jupyter Notebook. This visualization helps in understanding how the various elements of the domain model interact and are structured within the software architecture

```python


class_diagram = f"""
    Your task is to analyse Entities, Value Objects in detail and generate the following :
        1. Generate the class diagram for the Entities and Value Objects
        2. Give me diagram using mermaid as syntax in .md format 

    Service Context: {service_data}
"""

class_output = get_completion(class_diagram)
print(class_output)
```

# Entity-Relationship Diagram Generator from DDD Artifacts

ER diagrams are pivotal in designing and visualizing the database schema, showcasing the entities, their relationships, and the constraints that govern the data model. This automated process bridges the gap between domain modeling and database design, ensuring that the persistence layer aligns with the domain-centric approach advocated by DDD.

## Objective

The primary objective of generating ER diagrams from DDD artifacts is to translate the complex domain models into a structured database schema that accurately reflects the relationships and constraints of the domain entities. This visual representation aids developers, database designers, and architects in understanding and implementing the database schema that supports the microservices architecture.

## Workflow

1. **Input DDD Artifacts**: Review the Aggregates, Entities, and Value Objects identified for your bounded context to ensure they are accurately defined. These artifacts serve as the basis for generating the ER diagram.

2. **ER Diagram Generation**: Provide the detailed specifications of your DDD artifacts into the designated code cell. GPT will analyze these inputs to create an ER diagram that includes:
   - Entities representing both DDD Entities and Value Objects, with their attributes defined.
   - Relationships among entities, such as one-to-one, one-to-many, and many-to-many, reflecting the domain model's associations.
   - Key constraints and indices that are essential for the integrity and performance of the database.

3. **Visualization**: The generated ER diagram will be visualized within the Jupyter Notebook. This step helps in reviewing the proposed database schema, ensuring it is in harmony with the domain model and meets the requirements of the microservices architecture

```python
er_diagram = f"""
    Your task is to analyse Entities, Value Objects in detail and generate the following :
        1. Generate the Entity Relationship for the Entities and Value Objects
        2. Give me diagram using mermaid as syntax in .md format 

    Service Context: {service_data}
"""

er_output = get_completion(er_diagram)
print(er_output)
```

# OpenAPI Specification Generator

This section of the Jupyter Notebook leverages the Generative Pre-trained Transformer (GPT) to generate OpenAPI specifications for the microservices architecture previously defined. The OpenAPI Specification (OAS) is a widely adopted standard for describing RESTful APIs. It outlines the endpoints, operations (HTTP methods), parameters, and data models of an API, enabling clear communication about the capabilities of a service. This automated generation of the OpenAPI document is a critical step towards ensuring a consistent, understandable, and usable API interface that aligns with the domain-driven design and the underlying database schema.

## Goal

The goal of generating an OpenAPI specification is to provide a clear, detailed, and standardized documentation of the API that supports the microservices. This documentation is essential for developers working on or with the API, ensuring that they understand how to interact with it effectively. Additionally, the OpenAPI specification can be used to generate client libraries, server stubs, and API documentation tools, further streamlining the development process.

## Process

1. **Review Microservices and DDD Artifacts**: Before generating the OpenAPI specification, review the list of microservices, their bounded contexts, and the DDD artifacts (Aggregates, Entities, and Value Objects). Ensure that these elements are well-documented and clearly defined.

2. **Specification Generation**: Input the details of your microservices and DDD artifacts into the provided code cell. Upon execution, GPT will analyze this information to generate an OpenAPI specification. This specification will include:
   - API endpoints for each microservice, including paths and HTTP methods.
   - Request and response models for the API operations, based on the defined Entities and Value Objects.
   - Parameters, headers, and other necessary details for interacting with the API.
   - Security definitions, if applicable, to ensure secure access to the API.

3. **Visualization and Validation**: The generated OpenAPI specification will be output in either YAML or JSON format. You can visualize this specification using tools like Swagger UI or Redoc, which provide an interactive documentation interface. It is crucial to validate the generated specification with the development team and API consumers to ensure it meets all requirements and is accurately documented

```python
openapi_prompt = f"""
    Your task is to analyse service in detail and generate the following :
        1. Generate the API operations using RESTful API. Make sure that POST,PUT,DELETE,PATCH are completed for each resource.
        2. Make sure that all the attribute and operations are using snake_case syntax.
        3. Give detailed descriptions, summary and title when description are allowed.
        4. Put examples in each operation and schema, you can use faker to generate random data. Follow the field specification.
        5. Each resource should have your own tag.
        6. Give me full openapi in the last version in yaml format.

    Service Context: {service_data}
"""

api = get_completion(openapi_prompt)
print(api)
```
